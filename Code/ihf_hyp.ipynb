```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set();
```

```python
# Synthetic data (approximation)
bins = np.array([20, 40, 75, 150, 300, 500, 1000])
pca_vars = np.array([0.9, 0.99, 0.999, 0.9999, 0.99999])

# Left subplot (fixed PCA variance = 0.9999)
ihf_mah_ct = [0.58, 0.57, 0.60, 0.595, 0.59, 0.585, 0.59]
ihf_mah_mri = [0.22, 0.23, 0.15, 0.16, 0.125, 0.0905, 0.08]
ihf_nn_ct  = [0.48, 0.41, 0.415, 0.42, 0.415, 0.425, 0.42]
ihf_nn_mri = [0.49, 0.31, 0.30, 0.15, 0.145, 0.09, 0.091]

# Right subplot (fixed bins = 500)
ihf_mah_ct_pca = [0.86, 0.595, 0.59, 0.585, 0.61]
ihf_mah_mri_pca = [0.32, 0.13, 0.16, 0.075, 0.06]
ihf_nn_ct_pca  = [0.405, 0.405, 0.405, 0.425, 0.41]
ihf_nn_mri_pca = [0.11, 0.11, 0.115, 0.08, 0.12]

# Plot setup
fig, axes = plt.subplots(1, 2, figsize=(12, 4.5), sharey=True, dpi=200)

# Left panel
axes[0].plot(ihf_mah_ct, '-o', color='C0', label='IHF-Mah (CT)')
axes[0].plot(ihf_mah_mri, '-o', color='C1', label='IHF-Mah (MRI)')
axes[0].plot(ihf_nn_ct, '--^', color='C0', label='IHF-NN (CT)')
axes[0].plot(ihf_nn_mri, '--^', color='C1', label='IHF-NN (MRI)')

axes[0].set_xlabel('Number of bins $(m)$', fontsize=16)
axes[0].set_ylabel('mean FPR', fontsize=16)
axes[0].set_title('PCA explained variance = 0.9999', fontsize=16)
axes[0].legend(fontsize=14, loc='upper right')
axes[0].set_ylim(0, 1)

axes[0].set_xticks(range(len(bins)))
axes[0].set_xticklabels(bins)

# Right panel
axes[1].plot(ihf_mah_ct_pca, '-o', color='C0', label='IHF-Mah (CT)')
axes[1].plot(ihf_mah_mri_pca, '-o', color='C1', label='IHF-Mah (MRI)')
axes[1].plot(ihf_nn_ct_pca, '--^', color='C0', label='IHF-NN (CT)')
axes[1].plot(ihf_nn_mri_pca, '--^', color='C1', label='IHF-NN (MRI)')

axes[1].set_xlabel('PCA explained variance', fontsize=16)
axes[1].set_title('Number of bins = 500', fontsize=16)
axes[1].set_ylim(0, 1)

axes[1].set_xticks(range(len(pca_vars)))
axes[1].set_xticklabels(pca_vars)

# Increase tick label font size
for ax in axes:
    ax.tick_params(axis='both', labelsize=16)

plt.tight_layout()

plt.savefig("../Dissertation/Figures/5_ood_bench/ihf_hyp_gost.pdf", bbox_inches='tight')
#####plt.close()
# plt.show()
```
